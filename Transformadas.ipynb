```python
import matplotlib.pyplot as plt
from skimage import io, img_as_float
import numpy as np
```

TRANSFORMADA LOGARITMICA

```python
# Carregar a imagem
img = io.imread('log.png')

# Exibir a imagem
plt.imshow(img)
plt.axis('off')  # Remove os eixos
plt.show()
```

```python
# 1. Ler a imagem
I = io.imread('cameraman.tif')  # Certifique-se de que o arquivo 'cameraman.tif' está no diretório atual

# 2. Exibir a imagem original
plt.figure(figsize=(10, 8))  # Define o tamanho da figura
plt.subplot(2, 2, 1)
plt.imshow(I, cmap='gray',vmin=0, vmax=255)  # Exibe a imagem sem ajuste de escala
plt.title('Imagem Original')
plt.axis('off')


# 3. Converter a imagem para double (float no intervalo [0, 1])
Id = img_as_float(I)


# 4. Aplicar a transformação logarítmica com diferentes constantes
Output1 = 2*np.log(1+Id)
Output2 = 3*np.log(1+Id)
Output3 = 5*np.log(1+Id)

# 5. Exibir os resultados
plt.subplot(2, 2, 2)
plt.imshow(Output1, cmap='gray',vmin=0, vmax=1)  # Exibe a imagem sem ajuste de escala
plt.title('Output1: 2 * log(1 + Id)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(Output2, cmap='gray',vmin=0, vmax=1)  # Exibe a imagem sem ajuste de escala
plt.title('Output2: 3 * log(1 + Id)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(Output3, cmap='gray',vmin=0, vmax=1)  # Exibe a imagem sem ajuste de escala
plt.title('Output3: 5 * log(1 + Id)')
plt.axis('off')

# 6 Mostrar a figura
plt.tight_layout()
plt.show()
```

Transformada potencia (gamma)

Em geral, um valor de γ > 1 aumenta o contraste das regiões de valor alto da imagem em detrimento das regiões de valor baixo (transformação exponencial), enquanto observamos o oposto para γ < 1 . Isso confere à transformação de lei de potência propriedades semelhantes tanto à transformação logarítmica (γ < 1) quanto à transformação exponencial (γ > 1).

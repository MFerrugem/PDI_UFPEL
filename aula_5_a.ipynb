```python
# Importando os modulos
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, util,filters
from scipy.ndimage import convolve
```

```python
# Carregar a imagem
I = io.imread('eight.tif')
```

```python
# Criar figura com  3 subplots
plt.figure(figsize=(15,5))

# Imagem Original
plt.subplot(1,3,1)
plt.imshow(I, cmap='gray')
plt.title('Imagem Original')
```

```python
# Ruido sal e pimente /salt & pepper noise (3%)
Isp = util.random_noise(I, mode='s&p', amount=0.03)

# Imagem com rudo sal e pimenta
plt.subplot(1,3,2)
plt.imshow(Isp, cmap='gray')
plt.title('Salt & Pepper Noise')
```

```python
# Ruido gaussiano (variancia 0.02)
Ig = util.random_noise(I, mode='gaussian', var=0.02)

# Imagem com rudo gaussiando
plt.subplot(1,3,3)
plt.imshow(Ig, cmap='gray')
plt.title('Gaussian Noise')
```

```python
plt.subplot(1,3,1)
plt.imshow(I, cmap='gray')
plt.title('Imagem Original')
# Imagem com rudo sal e pimenta
plt.subplot(1,3,2)
plt.imshow(Isp, cmap='gray')
plt.title('Salt & Pepper Noise')

# Imagem com rudo gaussiando
plt.subplot(1,3,3)
plt.imshow(Ig, cmap='gray')
plt.title('Gaussian Noise')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Criar kernel de mdia
k = np.ones((3, 3)) / 9.0


# Aplicar filtro de mdia (scikit-image) - https://scikit-image.org/docs/0.24.x/auto_examples/filters/index.html
# Aplicar filtro de mdia (scipy)
Im = convolve(I, k)
Isp_m = convolve(Isp, k)
Ig_m = convolve(Ig, k)

# Plotar imagens
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(Im, cmap='gray')
plt.title('Imagem Original Filtrada')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(Isp_m, cmap='gray')
plt.title('Rudo Sal e Pimenta Filtrado')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(Ig_m, cmap='gray')
plt.title('Rudo Gaussiano Filtrado')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Histogram equalization

import matplotlib.pyplot as plt
import numpy as np
from skimage import io, img_as_ubyte
from skimage.exposure import equalize_hist

# Ler a imagem
I = io.imread('pout.tif')

# Converter a imagem para uint8 (necessrio para equalize_hist)
I = img_as_ubyte(I)

# Equalizao de histograma
Ieq = equalize_hist(I)

# Criar a figura com subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Mostrar as imagens
axes[0, 0].imshow(I, cmap='gray')
axes[0, 0].set_title('Imagem Original')
axes[0, 0].axis('off')

axes[0, 1].imshow(Ieq, cmap='gray')
axes[0, 1].set_title('Imagem Equalizada')
axes[0, 1].axis('off')

# Mostrar os histogramas
axes[1, 0].hist(I.ravel(), bins=256, range=(0, 256))  # Histograma da imagem original
axes[1, 0].set_title('Histograma Original')

axes[1, 1].hist((Ieq * 255).ravel(), bins=256, range=(0, 256))  # Histograma da imagem equalizada
axes[1, 1].set_title('Histograma Equalizado')

# Ajustar layout e mostrar o plot
plt.tight_layout()
plt.show()
```

O que est acontecendo?

1. A imagem 'pout.tif'  lida."
2. A imagem  convertida para uint8, pois a funo equalize_hist opera com esse tipo."
3. A equalizao de histograma  aplicada  imagem. Esta tcnica redistribui as intensidades de pixel para criar um histograma mais uniforme, aumentando o contraste."
4. As imagens original e equalizada so exibidas."
5. Os histogramas das imagens original e equalizada so exibidos.
 - O histograma da imagem original mostra a distribuio original das intensidades."
- O histograma da imagem equalizada mostra uma distribuio mais uniforme, indicando o aumento do contraste."
